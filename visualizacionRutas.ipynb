# Visualización de rutas

```python
import pandas as pd
import numpy as np
import folium
from folium.features import DivIcon
import osmnx as ox
import networkx as nx
```

```python
#ruta de vehiculo 0

ruta_vehiculo0 = pd.read_csv("rutas/100 clientes/Simulacion1/rutasVehiculo0.csv")
```

```python
ruta_vehiculo0.head()
```

```python


def desplegarRuta():

    try:
        vehicle_id = int(input("¿De qué vehículo quieres ver la ruta?: "))

    except:
        vehicle_id = -1
    vehiculos = {
        0:(ruta_vehiculo0["Latitud"],ruta_vehiculo0["Longitud"]),
    }

    colores = ["blue","red","yellow","orange","purple","green","brown"]

    
    if vehicle_id >= 0 and vehicle_id <= 6:

        coordenadas_vehicle = vehiculos[vehicle_id]

        G_walk = ox.graph_from_point((coordenadas_vehicle[0][0], coordenadas_vehicle[1][0]), dist=1000, network_type='walk')
        # initialize the road map from its starting point
        m = folium.Map(location=[coordenadas_vehicle[0][0], coordenadas_vehicle[1][0]],
                    zoom_start=13)

        #Visualize the path and specific points in the path (Sin incluir la bodega)
        latitudes = list(coordenadas_vehicle[0])[1:]
        longitudes = list(coordenadas_vehicle[1])[1:]

        points = []

        # read a series of points from coordinates and assign them to points object
        for i in range(len(latitudes)):
            points.append([latitudes[i], longitudes[i]])

        #Come back to the place of origin
        points.append(points[0])

        # specify an icon of your desired shape or chosing in place for the coordinates points
        for index,lat in enumerate(latitudes):
            if index == 0:

                folium.Marker([lat, 
                            longitudes[index]],
                            popup=('Deposito \n')
                            ,icon = folium.Icon(color='black',icon_color='white',prefix='fa', icon='shopping-cart')
                            ).add_to(m)
            else:
                folium.Marker([lat, 
                            longitudes[index]],
                            popup=('Bus Station{} \n '.format(index))
                            ,icon = folium.Icon(color=colores[vehicle_id],icon_color='white',prefix='fa', icon='car')
                            ).add_to(m)
        #Adding a pattern line to connect the dots
        folium.PolyLine(points,color="black",dash_array='5',opacity='0.85',tooltip="Transit Route 101").add_to(m)

        return m

    else:
        print("Ingresa un numero de vehiculo valido")
```

```python
desplegarRuta()
```
